```python
from itertools import product
```

```python
elementos = [
  {
    'nombre': "gafas",
    'beneficio': 50,
    'peso': 20
  },
  {
    'nombre': "cuaderno",
    'beneficio': 35,
    'peso': 25
  },
  {
    'nombre': "candela",
    'beneficio': 45,
    'peso': 20
  },
  {
    'nombre': "navaja",
    'beneficio': 20,
    'peso': 20
  },
  {
    'nombre': "agua",
    'beneficio': 35,
    'peso': 35
  },
  {
    'nombre': "sanduche",
    'beneficio': 40,
    'peso': 40
  },
  {
    'nombre': "cuerda",
    'beneficio': 50,
    'peso': 35
  },
  {
    'nombre': "carpa",
    'beneficio': 50,
    'peso': 30
  },
  {
    'nombre': "colchoneta",
    'beneficio': 45,
    'peso': 35
  },
  {
    'nombre': "sombrilla",
    'beneficio': 40,
    'peso': 25
  }
]
```

```python
def generar_combinaciones(n):
  return [list(comb) for comb in product([0, 1], repeat=n)]
```

```python
combinaciones = generar_combinaciones(len(elementos))
print(len(combinaciones))
print(combinaciones)
```

```python
posibilidades = []

for combinacion in combinaciones:
  peso = 0
  beneficio = 0
  for j in range(len(combinacion)):
    peso += combinacion[j] * elementos[j]['peso']
    beneficio += combinacion[j] * elementos[j]['beneficio']
  posibilidades.append({
    'combinacion': combinacion,
    'peso': peso,
    'beneficio': beneficio
  })
```

```python
for posibilidad in posibilidades:
  print(posibilidad)
```
